creer une fonction qui retourne le status de la requetes et l 'url créer

```python
def make_requests(url):
    r = requests.get(url)
    if r.ok:
        print(f"status ok {r.status_code}")
        return r.json()
    else:
        print(f"errors status {r.status_code}")
    print(r.url)
```

```python
def make_requests(url, payload):
    r = requests.get(url,params=payload)
    if r.ok:
        print(f"status ok {r.status_code}")
        print(r.url)
        return r.json()
    else:
        print(r.url)
        print(f"errors status {r.status_code}")
        
url_1 = "https://api.binance.com/api/v3/ticker/24hr"
payload_1 = {"symbol":"ADABNB"}
binance_req = make_requests(url_1, payload_1)
binance_req
```
